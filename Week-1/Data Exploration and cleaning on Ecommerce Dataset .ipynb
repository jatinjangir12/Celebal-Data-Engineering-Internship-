```python
print("\nStep 3: Handling Missing Values")

#identify the null values 
identify = df.isnull()
print(identify)
missing_before = df.isnull().sum()

print("Missing values per column:")
print(missing_before[missing_before > 0])
```

```python
print("\nFill missing discounts with 0")
df['discount'] = df['discount'].fillna(0)
```

```python
# Fill missing categorical metadata with a placeholder string
text_columns_to_fill = ['what_customers_said', 'seller_name', 'videos', 'seller_information', 'variations']
for col in text_columns_to_fill:
    df[col] = df[col].fillna('Not Available')

print(f"Total missing values remaining: {df.isnull().sum().sum()}")
```

```python
# STEP 4: Perform operations for filter rows and select columns
print("\nStep 4: Basic Filtering & Column Selection")

# Filter for highly-rated premium items (rating > 4.5)
premium_df = df[df['rating'] > 4.5]

# Select relevant columns for a clean summary view
summary_view = premium_df[['title', 'rating', 'ratings_count', 'initial_price', 'category']]
print(f"Filtered Rows (Rating > 4.5):", len(premium_df))
print("Sample Selection View:")
print(summary_view.head(3))
```

```python
print("\Step 5: Removing Duplicates")
duplicate_count = df.duplicated().sum()

print(f"Number of duplicate rows found:", duplicate_count)
df = df.drop_duplicates()
```

```python
print("\nStep 6: Feature Engineering (NEW Column)")
np.random.seed(42)
df['quantity'] = np.random.randint(1, 6, size=len(df))

```

```python
#element-wise vectorize multiplication for the new column
df['total_amount'] = df['initial_price'] * df['quantity']

print("Sample total_amount column:")

print(df[['title', 'initial_price', 'quantity', 'total_amount']].head())
```

```python

df['total_amount'] = df['initial_price'] * df['quantity']

selected_columns = ['title', 'initial_price', 'quantity', 'total_amount']

df_summary = df[selected_columns]

print("Summary File for title, initial_price, quantity, total_amount ")
print(df_summary.head(5))

output_summary_file = 'product_totals_summary.csv'
df_summary.to_csv(output_summary_file, index=False)

print(f"\n A separate file for inital_price, quantity, total_amount after removing null values and applying Feature Engineering: '{output_summary_file}'")
```

```python
output_csv_filename = 'Combined_dataset_final_cleaned.csv'

df.to_csv(output_csv_filename, index=False)

print(f"\nStep 7: Cleaned Dataset Saved to '{output_csv_filename}' ")
```
